#Label processing


```python
class Config:

  test_dir  = '/content/drive/MyDrive/yolov5/data/MUJEEB'
```

```python
def read_images(folder_path):

  # List to store image file names
  image_list = []

  # Loop through all files in the folder
  for filename in os.listdir(folder_path):
      if filename.endswith(('.png', '.jpg', '.jpeg')):
          image_list.append(filename)

  return image_list
```

```python
def read_file_to_list(file_path):
    # Open the file and read it line by line
    with open(file_path, 'r') as file:
        # Read all lines and strip newline characters
        lines = [line.strip() for line in file.readlines()]
    return lines
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
from PIL import Image
import pandas as pd
import numpy as np

def plot_images_with_bboxes_from_df(df, num_images):
    # Sample random images from the dataframe
    sampled_df = df.sample(n=min(num_images*num_images, len(df)))

    grid_size = int(num_images)

    # Create a figure and set of subplots
    fig, axs = plt.subplots(grid_size, grid_size, figsize=(15, 15))

    # Ensure axs is always treated as an array
    axs = np.array(axs).flatten()

    # Loop through the sampled images and plot them with bounding boxes
    for idx, (_, row) in enumerate(sampled_df.iterrows()):
        image_path = f'{Config.train_dir}images/{row["image"]}'
        labels = row["label"]

        # Open the image
        img = Image.open(image_path)
        img_width, img_height = img.size

        # Display the image on the current subplot
        axs[idx].imshow(img)

        # Process labels
        for label in labels:
            # Parse each label: class_id, x_center, y_center, width, height
            class_id, x_center, y_center, bbox_width, bbox_height = map(float, label.split())

            # Convert normalized coordinates back to image dimensions
            x_center *= img_width
            y_center *= img_height
            bbox_width *= img_width
            bbox_height *= img_height

            # Calculate the top-left corner of the bounding box
            top_left_x = x_center - bbox_width / 2
            top_left_y = y_center - bbox_height / 2

            # Create a rectangle patch for the bounding box
            rect = patches.Rectangle((top_left_x, top_left_y), bbox_width, bbox_height, linewidth=2, edgecolor='r', facecolor='none')

            # Add the rectangle to the plot
            axs[idx].add_patch(rect)

            # Add the class label above the bounding box
            axs[idx].text(top_left_x, top_left_y - 5, f'Class: {int(class_id)}', color='yellow', fontsize=12, backgroundcolor='black')

        # Remove axes for a cleaner look
        axs[idx].axis('off')

    # Turn off the extra subplots if fewer than 9 images are plotted
    for i in range(idx + 1, grid_size * grid_size):
        axs[i].axis('off')

    # Display the grid
    plt.tight_layout()
    plt.show()
```

```python
test_images = read_images(f'{Config.test_dir}')
```

```python
test_df = pd.DataFrame(test_images, columns=['image'])
test_df['label'] = test_df['image'].apply(lambda x: read_file_to_list(f"{Config.test_dir}{x.replace('.jpg', '.txt')}"))
test_df["label_count"] = test_df["label"].apply(lambda x: len(x))

```

```python
# @title                epoch vs       train/cls_loss

from matplotlib import pyplot as plt
import seaborn as sns
def _plot_series(series, series_name, series_index=0):
  palette = list(sns.palettes.mpl_palette('Dark2'))
  xs = series['               epoch']
  ys = series['      train/cls_loss']

  plt.plot(xs, ys, label=series_name, color=palette[series_index % len(palette)])

fig, ax = plt.subplots(figsize=(10, 5.2), layout='constrained')
df_sorted = history.sort_values('               epoch', ascending=True)
_plot_series(df_sorted, '')
sns.despine(fig=fig, ax=ax)
plt.xlabel('               epoch')
_ = plt.ylabel('      train/cls_loss')
```
